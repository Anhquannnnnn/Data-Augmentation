```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Gnrer des donnes artificielles
np.random.seed(0)
X1 = np.random.normal(loc=[0, 0], scale=0.5, size=(100, 2))
X2 = np.random.normal(loc=[4, 4], scale=0.5, size=(100, 2))
X3 = np.random.normal(loc=[0, 4], scale=0.5, size=(100, 2))

# Regrouper les donnes
X = np.vstack([X1, X2, X3])

# Crer et ajuster un modle GaussianMixture avec 3 clusters
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm.fit(X)

# Prdire les tiquettes de clusters
labels = gmm.predict(X)

# Afficher les clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)
plt.title('Clustering par Gaussian Mixture (3 clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```
